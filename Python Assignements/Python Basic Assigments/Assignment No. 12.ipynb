**`Run-Object :-`** In our docs file we have lot of structure and any document object contains list of paragraphs objects.In every paragraphs we have some run objects like some words in bold,intalic and some in underline and tab as well.
So these objects count as Run Objects.

**`Paragraph Object :-`** At the highest level, a Document object represents the entire document. The Document object contains a list of Paragraph objects for the paragraphs in the document.
Each of these Paragraph objects contains a list of one or more Run objects. 

Inside the paragraph object we have space,text and allignment are count.

If u see the above complete is the paragraph object and each text has its own properties like text and its italic,underline and normal text are called as Run-Object.

These are the Run type objects.

- Many font properties are tri-state, 
- Meaning they can take the values **`True`**, **`False`**, and **`None`**. 
- **`True`** means the property is **on** 
- **`False`** means it is **off**.
- Conceptually, the **`None`** value means **`inherit`**.

```python
import docx
doc = docx.Document() 
doc.add_heading('GeeksForGeeks', 0)
doc.add_paragraph("Coronavirus disease (COVID-19) is an infectious disease caused by a newly discovered coronavirus. Most people who fall sick with COVID-19 will experience mild to moderate symptoms and recover without special treatment.") 
doc.save('New.docx')
```

```python
import docx
doc = docx.Document('New.docx')
doc.add_paragraph('Hello,there!')
doc.save('NewOne.docx')
```

- When the level is set to `0`, the string is printed as the title of the document.
- For all other values it prints a heading. The size of heading decreases as the level increases. If no level is set, by default its value is always `1`.
- ValueError if any value other than from this range is given as input.
