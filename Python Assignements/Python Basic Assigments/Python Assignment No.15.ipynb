```python
def isPrime(num):
    for i in range(2,num):
        if (num%i)==0:
            return False
    else:
        return True
```

```python
def primeGenerator(n):
    lst = []
    num=2
    while n>1:
        if isPrime(num):
            yield num
            n-=1       #once we acheive n value we have to decrease it
        num+=1         #in order to avoid to check wheather
    return lst        
```
