```python
import re
import os

lst = []

file = str(os.listdir())
pattern = re.compile(r'[A-Za-z0-9.() _&*!]+\.[a-zA-Z]+')
matches = pattern.finditer(file)


for match in matches:
    lst.append(match.group())
    

os.mkdir(os.getcwd()+'\\ExtensionsFiles')

import shutil

for i in lst:
    shutil.copy('F:\\Programming\\Full Stack Data Science\\Python Programming Language\\Python Assignments\\{}'.format(i),
                'F:\\Programming\\Full Stack Data Science\\Python Programming Language\\Python Assignments\\ExtensionsFiles')
```
