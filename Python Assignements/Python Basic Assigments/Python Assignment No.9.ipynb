The relative path is noting but path which is related to our present directory.Relative path is defined as the path related to the present working directly(pwd). It starts at your current directory

- Absolute path is a complete path from start of actual file system from / directory.
- An absolute path refers to the complete details needed to locate a file or folder, starting from the root element and ending with the other subdirectories.
- Absolute paths are used in websites and operating systems for locating files and folders.
- An absolute path is also known as an absolute pathname or full path.

The readlines method returns the contents of the entire file as a list of strings, where each item in the list represents one line of the file.
- 	It Returns a list of strings, each representing a single line of the file. If n number of charecters is not provided then all lines of the file are returned.
