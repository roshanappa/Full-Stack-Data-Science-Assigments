```python
guess_me = int(input('Please enter the number 7 :-',))

if guess_me >7:
    print('too high')
elif guess_me<7:
    print('too low')
elif guess_me==7:
    print('just right')
else:
    print('Sorry !!')
```

```python
guess_me = 7
start =1
i=1
while i>0:
    if start < guess_me:
        print('too low')
    elif start > guess_me:
        print('oops')
    elif start==guess_me:
        print('found it !!')
    break
    i+=1
```

```python
def get_odds(n=10):
    for i in range(n):
        if i%2!=0:
            yield i
```

```python
for i,j in enumerate(get_odds()):
    if i ==3:
        print(j)
```

In general, when a Python script encounters a situation that it cannot cope with, it raises an exception. An exception is a Python object that represents an error.

- Syntax error 
- Run-time eror or compile error.
