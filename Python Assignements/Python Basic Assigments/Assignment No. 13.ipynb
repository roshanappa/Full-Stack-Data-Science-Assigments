```python
from csv import reader,writer
```

```python
with open('Text.csv', newline='') as csvfile:
    csv_reader = reader(csvfile,delimiter=' ',quotechar='|')
    for line in csv_reader:
        print(' '.join(line),end='')
```

```python
with open('Text.csv', mode='w',newline='') as csvfile:
    csv_writer = writer(csvfile,delimiter=' ')
    csv_writer.writerow(['I am very to learn new function'])
    csv_writer.writerow(['This is wonderful.'])
```

```python
file = open('Text.csv','r')
for i in file.readlines():
    print(i,end='')
```

- For the reader we required **`r`** mode to read the file
- For the writer we required **`w`** mode to write the something in file.
