```python
import string
lst = ['mozzarella','cinderella','salmonella']
new_list = []
for i in range(len(lst)):
    word=string.capwords(lst[i])
    new_list.append(word)
print(new_list)
```

```python
import string
lst = ['mozzarella','cinderella','salmonella']
new_list = []
for i in range(len(lst)):
    word=str.capitalize(lst[i])
    new_list.append(word)
print(new_list)
```
