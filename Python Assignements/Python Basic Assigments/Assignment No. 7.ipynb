In standard expressions,we use this charecter for `bitwise OR operator` in python syntax.

It is use in regular expression of conditional purpose,whenever we have mutiple options and we want use that options as per the our expression and condtion in that time we gonna use the `|` this charecter.

```python
import re
greeting = '''

Hi Geeta
Hello Geeta
Good Morning Geeta
ssddddldld 
hdhddkjkd 
dhdddldhn 
jjddk 
Roshan

'''
pattern = re.compile(r'(Hi|Hello|Good Morning)\s[A-Za-z0-9]+')
matches = pattern.finditer(greeting)

for match in matches:
    print(match.group())
```

#### `+ Charecter`

**+ Charecter** will gives us the exact match string in one iteration like below way.

```python
string = 'I am looking for Data Scientist Job,job is not ready in market for fresher.'
pattern = re.compile(r'[a-bA-B]+')
matches = pattern.finditer(string)

for i in matches:
    print(i)
```

**`*`Charecter** will not gives us extact match of the  string in one iteration like below way.

```python
string = 'I am looking for Data Scientist Job,job is not ready in market for fresher.'
pattern = re.compile(r'[a-bA-B]*')
matches = pattern.finditer(string)

for i in matches:
    print(i)
```

**`{4}`**:- It is going to represent the extact number charecters we want the string.

```python
test_string = 'Hi I have 200000 rupees.'
pattern = re.compile(r'\d{4}')
matches = pattern.finditer(test_string)

for match in matches:
    print(match)
    print(match.group())
```

It will return the extact number from the left to right of the text.

**`{4,5}`**:- It is going to represent the range of charecters we want the string.
- If u can see that I wanted the digits from 1 to 6 and I got that


```python
test_string = 'Hi I have 200000 rupees.'
pattern = re.compile(r'\d{1,6}')
matches = pattern.finditer(test_string)

for match in matches:
    print(match.group())
```

- If I want to get the number from 1 to 4 then I can find it.

```python
test_string = 'Hi I have 200000 rupees.'
pattern = re.compile(r'\d{1,4}')
matches = pattern.finditer(test_string)

for match in matches:
    print(match.group())
```

- **`\d`** :- This is basically used to extract the digits or numerical sub-string from the main-string
- **`\w`**:-This is basically used to extarct the alpha-numeric charecters from mainp-string.
- **`\s`**:- This is used to define the white-space charecter or maintain the space inside the sub-string from the main-string.

- **`\D`** :- This is basically not used to extract the digits or numerical sub-string from the main-string
- **`\W`**:-This is basically not used to extarct the alpha-numeric charecters from mainp-string.
- **`\S`**:- This is not used to define the white-space charecter or not maintain the space inside the sub-string from the main-string.

- **`.`**:- It is helping to the any chareter inside the matched string but not for the new-line.
- **`*`**:- It is helping us to get the occurance of the substring or match string from zero to many.Its start to find the substring from zero index to many index.
- **`?`**:- It is used to suuggest optional charecters inside the match string.It also used at the end of that charecter which we want keep as optional.

```python
string = 'ANDDJDSHSJa5HDSHDHDLAHDD45454545i4689ld'
pattern = re.compile(r'[0-9a-z]+')
matches = pattern.finditer(string)

for match in matches:
    print(match.group())
    print(type(match))
    print()
```

```python
import re
string = 'I am living in my Dream !!'
pattern  = re.compile(r"dream",re.IGNORECASE)
matches = pattern.finditer(string)

for match in matches:
    print(match.group())
```

This is how we can do the insensetive regex case.

**`.`** charecter is normally match the any charecter before and after after the selected substring but that should be one line.

```python
import re
string = 'I am living in my Dream !!'
pattern  = re.compile(r"..dream",re.IGNORECASE)
matches = pattern.finditer(string)

for match in matches:
    print(match.group())
```

Here I have got the two letter before the Dream **`sub-string`** comes inside the **`main-string`**.

```python
import re
string = """ I am living in my Dream !!.
             I love dream."""
pattern  = re.compile(r"dream",re.DOTALL)
matches = pattern.finditer(string)

for match in matches:
    print(match)
```

- I have used the regex with second attribute as **`re.DOTALL`**
- It allow us to get the match of substring with new line as well unlike **`.`**.

#### Between the commas

```python
test_string = """42

1,234
6,368,745
65,678,8889,99,88
55
67
100
500
77
"""

import re
pattern = re.compile(r'\,\d{3}\,')
matches = pattern.finditer(test_string)

for match in matches:
    print(match)
```

#### Between the commas

```python
test_string = """42

1,234
6,368,745
65,678,8889,99,88
55
67
100
500
77
"""

import re
pattern = re.compile(r'\,?\d{3}\,?')
matches = pattern.finditer(test_string)

for match in matches:
    print(match)
```
