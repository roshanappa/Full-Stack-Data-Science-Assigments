```python
import re
eggs = input('Enter your favorite string1 :-')
bacon = input('Enter your favorite string2 :-')

regex1= re.compile(r'[a-zA-Z]+',re.IGNORECASE)
regex2= re.compile(r'[a-zA-Z]+',re.IGNORECASE)

match1 = regex1.match(eggs)
match2 = regex1.match(bacon)

assert match1.group()==match2.group()
assert match1.group() != match2.group()
```

**`Step In :-`** It is used to deep dive inside the code which is avaliable in our IDE.like suppose we have for loop for the range(10) and if we do the step in means it will check for every iteration what kind of output we are getting.It shows hows our code is getting flow step by step.i.e.If we have function it will deep dive inside the function return everything whichever the code avaliable inside that and then  move to move to the next line of code.

Steps into the method to show what happens inside it. Use this option when you are not sure the method is returning a correct result.

**`Step Over :-`** By using the breakpoint inside the pycharm we will step over the code so that everytime will not debugg the code in deep dive,It just use new line everytime.If any comes under the line then it will execute its result instead of going deep dive inside that.

Steps over the current line of code and takes you to the next line even if the highlighted line has method calls in it. The implementation of the methods is skipped, and you move straight to the next line of the caller method

**`Step Out :-`** Steps in the method to show what happens inside any function but after finding that the result is good then we have step out from the code and during this period we will gonna use the step out debugger. 

Stepping out skips all iterations of the loop and takes you straight to the main method (the caller).Steps out of the currently executed code block. This lets you skip the remaining iterations of while, for and alike without exiting the enclosing method.

It stop proceding debugging whenever it reaches a breakpoint.

- Inside the IDE,we have breakpoint in order to check bugg inside code for selected line of code.
- It also help to get the rid from excecution of code after reaching the breakpoint of IDE.
- It is also helpful when we have the 1000 number iteration inside the for loop and we want to detect the iteration at which the code is getting issue for that that we will gonna use the conditional breakpoint.
- In software development, a breakpoint is an intentional stopping or pausing place in a program, put in place for debugging purposes. It is also sometimes simply referred to as a pause. More generally, a breakpoint is a means of acquiring knowledge about a program during its execution.
