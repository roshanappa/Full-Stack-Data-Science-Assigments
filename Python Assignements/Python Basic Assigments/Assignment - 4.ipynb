Suppose we have list like below

```python
lst = ['122',23,5]
lt = lst
print(id(lst))
print(id(lt))
```

In that it having some allocation of memory and with same address 

If any list varible data will cahnge then it will affect on the other one as well.

```python
lst[0]=1   #they both are getting change and adres also the same 
print(lt)
print(lst)
print(id(lt))
print(id(lst))
```

That is shollow copy if we change any variable if will affect on other variable variable as well.

But in deep it is exactly opposite and not allows to change paramter when other will change

```python
sa = [24,55,67,8]
```

```python
da = sa.copy()
```

```python
sa[1]=20
print(sa) #here we can see one variable change the value but other was not change thier value because we have done deepcopy.
print(da)
```

Shollow copy is the example of call by value and deep copy is teh example of call by reference.
