### Question 10 :-

#### break :-

- It is initiated in the loop,when we want to break the loop at the particular iteration in other way it help us to simply break or stop iterating over the sequence of data and also It do not run further code.
- For stoping an itration we have to mention the condition when that condition is getting True then will break loop and come out of loop.

```python
i = 1
while i < 10:
  print('Hello')
  if i == 3:
    break
  i += 1
```

**continue :-**

- It is used to skip the particular instances from the loop like if we do not any number which is fully divisional by 2 then we can simply say continue after writting the condition and print the statement.

- It means it help us to skip that particular iteration while itearting throgh the itearable object.

```python
for i in range(10):
    if i%2==0:
        continue
    print(i)
```
